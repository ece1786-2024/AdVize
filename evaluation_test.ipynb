**Notes for resutls: **
- Precision: measures how much of the generated query is relevant to the persona text
- Recall: measures how much the persona text is covered by the generated query
- F1: harmonic mean of the precision and recall (F1 = 2 * ((precision * recall)/(precision + recall)))

```python
from bert_score import score

persona_references = [
    "Basketball",
    "Soccer",
    "Fashion",
    "Travel",
    "Technology"
]

generated_queries = [
    "What are the best AI tools for developers?",
    "What are some popular cat toys",
    "Give me some nice kitchen tools"
]

print("\nQuery relevance scores using BERTScore with multiple references:")
for query in generated_queries:
    P, R, F1 = score([query] * len(persona_references), persona_references, lang='en', verbose=True)

    # maximum F1 score across all references
    max_f1 = max(F1).item()
    print(f"Query: '{query}'\nMax F1 score across references: {max_f1:.4f}\n")
```

Revised Approach

- Extract key phrases from the query using KeyBERT.
- Use semantic similarity between the extracted key phrases and the persona keywords to assign a relevance score.
- Calculate a similarity score between each extracted key phrase and the persona keywords directly, without needing any broad topic mapping.

```python
from keybert import KeyBERT
from sentence_transformers import SentenceTransformer, util
import numpy as np

kw_model = KeyBERT('all-MiniLM-L6-v2')
sentence_model = SentenceTransformer('all-MiniLM-L6-v2')

persona_keywords = ["basketball", "soccer", "fashion", "travel", "technology"]

query = "What is the newest released Nike shoe?"

key_phrases = kw_model.extract_keywords(query, keyphrase_ngram_range=(1, 2), stop_words='english', top_n=5)
extracted_phrases = [phrase[0] for phrase in key_phrases]
print(f"Extracted Key Phrases: {extracted_phrases}")

persona_embeddings = {keyword: sentence_model.encode(keyword, convert_to_tensor=True) for keyword in persona_keywords}
phrase_embeddings = {phrase: sentence_model.encode(phrase, convert_to_tensor=True) for phrase in extracted_phrases}

generalized_topics = []
for phrase, phrase_embedding in phrase_embeddings.items():
    max_similarity = 0
    best_match = None
    for keyword, persona_embedding in persona_embeddings.items():
        similarity = util.cos_sim(phrase_embedding, persona_embedding).item()
        if similarity > max_similarity:
            max_similarity = similarity
            best_match = keyword
    generalized_topics.append(best_match)

print(f"Generalized Topics for query '{query}': {generalized_topics}")

scores = []
for topic in generalized_topics:
    if topic:
        topic_embedding = sentence_model.encode(topic, convert_to_tensor=True)
        query_embedding = sentence_model.encode(query, convert_to_tensor=True)
        similarity_score = util.cos_sim(query_embedding, topic_embedding).item()
        scores.append(similarity_score)

if scores:
    average_score = np.mean(scores)
else:
    average_score = 0.0

print(f"Relevance Score for query '{query}': {average_score:.4f}")
```

```python
from keybert import KeyBERT
from sentence_transformers import SentenceTransformer, util
import numpy as np

kw_model = KeyBERT('all-MiniLM-L6-v2')
sentence_model = SentenceTransformer('all-MiniLM-L6-v2')

persona_keywords = ["AI", "sports", "fashion", "travel"]

query = "What is the newest released Nike shoe?"

# extract key phrases using KeyBERT
key_phrases = kw_model.extract_keywords(query, keyphrase_ngram_range=(1, 2), stop_words='english', top_n=5)
extracted_phrases = [phrase[0] for phrase in key_phrases]
print(f"Extracted Key Phrases: {extracted_phrases}\n")

# convert key phrases and persona keywords into embeddings
extracted_embeddings = {phrase: sentence_model.encode(phrase, convert_to_tensor=True) for phrase in extracted_phrases}
persona_embeddings = {keyword: sentence_model.encode(keyword, convert_to_tensor=True) for keyword in persona_keywords}

# not using WordNet..
# find related terms using cosine similarity within the extracted phrases and persona interests
def find_similar_terms(embedding, all_embeddings, top_n=3):
    similarities = []
    for term, emb in all_embeddings.items():
        similarity = util.cos_sim(embedding, emb).item()
        similarities.append((term, similarity))
    # Sort by similarity and return the top n terms
    similarities.sort(key=lambda x: x[1], reverse=True)
    return similarities[:top_n]

# broaden key phrases using embeddings
broadened_query_topics = set()
for phrase, phrase_embedding in extracted_embeddings.items():
    similar_terms = find_similar_terms(phrase_embedding, persona_embeddings, top_n=3)
    for term, score in similar_terms:
        broadened_query_topics.add(term)
    print(f"Broadened terms for '{phrase}': {[term for term, score in similar_terms]}")

print(f"\nBroadened Query Topics: {broadened_query_topics}\n")

# broaden the persona keywords using embeddings
broadened_persona_topics = set()
for keyword, persona_embedding in persona_embeddings.items():
    similar_terms = find_similar_terms(persona_embedding, extracted_embeddings, top_n=3)
    for term, score in similar_terms:
        broadened_persona_topics.add(term)
    print(f"Broadened terms for persona keyword '{keyword}': {[term for term, score in similar_terms]}")
broadened_persona_topics.update(persona_keywords)

print(f"\nBroadened Persona Topics: {broadened_persona_topics}\n")

print("Similarity Scores:")
relevant = False
for query_topic in broadened_query_topics:
    for persona_topic in broadened_persona_topics:
        query_emb = sentence_model.encode(query_topic, convert_to_tensor=True)
        persona_emb = sentence_model.encode(persona_topic, convert_to_tensor=True)
        similarity = util.cos_sim(query_emb, persona_emb).item()
        print(f"Similarity between '{query_topic}' and '{persona_topic}': {similarity:.4f}")
        if similarity > 0.6:
            print(f"Matched: '{query_topic}' with Persona Topic: '{persona_topic}' (Similarity: {similarity:.4f})")
            relevant = True
            break
    if relevant:
        break

print(f"\nRelevance: {'Relevant' if relevant else 'Not Relevant'}")
```

```python
import requests
from keybert import KeyBERT
from sentence_transformers import SentenceTransformer, util

kw_model = KeyBERT('all-MiniLM-L6-v2')
sentence_model = SentenceTransformer('all-mpnet-base-v2')

persona_keywords = ["AI", "sports", "fashion", "travel"]

query = "What is the newest released Nike shoe?"

key_phrases = kw_model.extract_keywords(query, keyphrase_ngram_range=(1, 2), stop_words='english', top_n=5)
extracted_phrases = [phrase[0] for phrase in key_phrases]
print(f"Extracted Key Phrases: {extracted_phrases}\n")

def broaden_with_conceptnet(term):
    url = f"http://api.conceptnet.io/c/en/{term}?limit=5"
    response = requests.get(url)
    related_terms = set()
    if response.status_code == 200:
        data = response.json()
        for edge in data.get('edges', []):
            related_term = edge['end']['label'].lower()
            if related_term != term:
                related_terms.add(related_term)
    return list(related_terms)

broadened_query_topics = set()
for phrase in extracted_phrases:
    words = phrase.split()
    for word in words:
        broader_terms = broaden_with_conceptnet(word)
        print(f"Broadened terms for '{word}': {broader_terms}")
        broadened_query_topics.update(broader_terms)

print(f"\nBroadened Query Topics: {broadened_query_topics}\n")

query_embeddings = {topic: sentence_model.encode(topic, convert_to_tensor=True) for topic in broadened_query_topics}
persona_embeddings = {keyword: sentence_model.encode(keyword, convert_to_tensor=True) for keyword in persona_keywords}

print("Similarity Scores:")
relevant = False
for query_topic, query_emb in query_embeddings.items():
    for persona_topic, persona_emb in persona_embeddings.items():
        similarity = util.cos_sim(query_emb, persona_emb).item()
        print(f"Similarity between '{query_topic}' and '{persona_topic}': {similarity:.4f}")
        if similarity > 0.6:
            print(f"Matched: '{query_topic}' with Persona Topic: '{persona_topic}' (Similarity: {similarity:.4f})")
            relevant = True
            break
    if relevant:
        break

print(f"\nRelevance: {'Relevant' if relevant else 'Not Relevant'}")
```
